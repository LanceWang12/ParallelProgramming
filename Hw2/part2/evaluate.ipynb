```python
import matplotlib.pyplot as plt
import numpy as np
```

## Q3

```python
def plot(record):
    ideal = np.arange(1, 1 + len(record))
    plt.figure(figsize = (6, 4), dpi = 100)
    plt.title('Speed up rate in different numThreads')
    plt.plot(ideal, ideal, label = 'ideal', marker = 'o')
    plt.plot(ideal, record, label = 'real', marker = 'o')
    plt.xticks(ideal)
    plt.xlabel('Thread num')
    plt.ylabel('Execution time (sec)')
    plt.legend()
    plt.show()

def bar_plt(record1, record2, width = 0.35):
    thread_id = [i for i in range(len(record1))]
    thread_name = [f'Thread{ID}' for ID in thread_id]
    plt.figure(figsize = (6, 4), dpi = 100)
    plt.title('Execution time about every thread')
    plt.bar(thread_id, record1, width = width, label = 'origin')
    plt.bar(np.array(thread_id) + width, record2, width = width, label = 'optim')
    plt.xticks(np.array(thread_id) + width / 2, thread_name)
    plt.xlabel('Thread ID')
    plt.ylabel('Execution time (sec)')
    plt.legend()
    plt.show()
```

```python
# view1
record = [1, 1.96, 2.92, 3.81]
plot(record)

# view2 
record = [1, 1.98, 2.96, 3.84]
plot(record)
```

```python
# view1
# 4 thread
record1 = [0.0405691, 0.17266, 0.173636, 0.0410244]
record2 = [0.108017, 0.108259, 0.108519, 0.108133]
bar_plt(record1, record2)

# 5 thread
record1 = [0.0214529, 0.109544, 0.168363, 0.109954, 0.021635]
record2 = [0.0930495, 0.0864632, 0.0863985, 0.0862982, 0.0929336]
bar_plt(record1, record2)
```

```python
# view2
# 4 thread
record1 = [0.0949936, 0.053116, 0.0535871, 0.0512728]
record2 = [0.0685573, 0.0625472, 0.0621822, 0.0678058]
bar_plt(record1, record2)

# 5 thread
record1 = [0.0852537, 0.0422394, 0.0464055, 0.0410236, 0.0454246]
record2 = [0.0557675, 0.0510494, 0.0505132, 0.0502567, 0.0547975]
bar_plt(record1, record2)
```
